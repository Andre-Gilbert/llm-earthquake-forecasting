```python
import pandas as pd
from datetime import timedelta
from tqdm import tqdm
```

```python
start_date = pd.Timestamp('1924-01-01')
end_date = pd.Timestamp('1924-01-31')
date_ranges = [(start_date.date(), end_date.date())]

while end_date < pd.Timestamp('2024-05-31'):
    start_date = end_date + timedelta(days=1)
    end_date = start_date + timedelta(days=15)
    date_ranges.append((start_date.date(), end_date.date()))

if date_ranges[-1][1] > pd.Timestamp('2024-05-31').date():
    date_ranges[-1] = (date_ranges[-1][0], pd.Timestamp('2024-05-31').date())
```

```python
df = pd.DataFrame()
for start_time, end_time in tqdm(date_ranges):
    url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=csv&starttime={start_time}&endtime={end_time}&eventtype=earthquake&limit=20000&minmagnitude=-1"
    temp_df = pd.read_csv(url)
    df = pd.concat([df, temp_df], ignore_index=True)
```
